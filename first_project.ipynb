```python
new_data = """
        SELECT movie_basics.start_year, movie_basics.runtime_minutes, movie_ratings.averagerating,movie_ratings.numvotes
        FROM movie_basics
        JOIN movie_ratings ON movie_basics.movie_id = movie_ratings.movie_id
        """

# Execute the query and fetch the results
results_idbm = conn.execute(new_data).fetchall()
results_idbm
```

Check for null values in our dataset.

```python
results_idbm = pd.DataFrame(results_idbm)
results_idbm
```

```python
corr = results_idbm[1].corr(results_idbm[2])
print(corr)
```

This means that movies with longer run times are slightly less likely to receive high average ratings, while movies with shorter run times are slightly more likely to receive high average ratings. However, the strength of this association is weak, which means that it may not be very meaningful or useful for predicting or analyzing trends in the data.
Note that correlation does not necessarily imply causation, and that other factors or variables may also influence the relationship between these two columns.
